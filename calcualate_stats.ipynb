```python
import pandas as pd 

import os 
```

```python
os.listdir(".")
```

```python
stats_df = pd.DataFrame(columns = ["Dataset", "Rows", "Columns",  "Missing_values", "Classes"])

i = 0
for dataset in os.listdir("."):
    if os.path.isfile(f'./{dataset}') or dataset == '.git':
        continue
    df = pd.read_parquet(f"./{dataset}/{dataset}.parquet")

    stats_df.loc[i, "Dataset"] = dataset
    stats_df.loc[i, "Rows"] = df.shape[0]
    stats_df.loc[i, "Columns"] = df.shape[1]
    stats_df.loc[i, "Missing_values"] = df.isnull().values.any()
    stats_df.loc[i, "Classes"] = len(df["class"].unique())
    i+=1

```
